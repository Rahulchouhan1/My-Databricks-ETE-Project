```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

##  **Read Stream For Orders Data**

```python
orders = spark.readStream.format("cloudFiles") \
    .option("cloudFiles.format", "parquet") \
    .option("cloudFiles.schemaLocation", f"/Volumes/sampleproject/practice/bronze/checkpoint_orders") \
    .load(f"/Volumes/sampleproject/practice/source/orders")
```

### **Write Orders Data With AutoLoader**

```python
orders.writeStream.format("parquet") \
    .outputMode("append") \
    .option("checkpointLocation", f"/Volumes/sampleproject/practice/bronze/checkpoint_orders") \
    .option("path", f"/Volumes/sampleproject/practice/bronze/orders") \
    .trigger(availableNow=True) \
    .start()
```
